```python
import numpy as np
def solubility( temperature: float):
    A1 = -3.15
    A2 = -2263.75
    A3 = 9.94
    A4 = -4755.72
    A5 = 12474.32
    A6 = -14846.20
    A7 = 6098.95
    TK = temperature + 273.15
    x0cal = np.exp(A1 + A2 / TK + A3 + (A4+A5+A6+A7)/TK)
    return x0cal
```
