```python
solubility_unstable = lambda T: (8.437*10**-3*T**2 + 3.032*10**-2*T + 4.564)/1000 # degC kg/kg
solubility_stable = lambda T: (7.644*10**-3*T**2 - 1.165*10**-1*T + 6.222)/1000 # degC kg/kg
```

```python
from plotly import graph_objects as go
import numpy as np


temperature_range = np.linspace(10,60)
c_stable = solubility_stable(temperature_range)
c_unstable = solubility_unstable(temperature_range)

fig = go.Figure([
    go.Scatter(x=temperature_range, y=c_unstable, name="unstable"),
    go.Scatter(x=temperature_range, y=c_stable, name="stable"),
])

fig.update_layout(
    yaxis={"title": "Concentration (kg/kg)"}, 
    xaxis={"title": "Temperature ($^\circ C$)"},
    showlegend=True,
)
fig
```

```python
# Calculator
volume = 200 * 1e-6 # m3
density = 998 # kg/m3

water_mass = volume * density # kg
crystal_mass = water_mass * solubility_unstable(40)
crystal_mass_final = water_mass * solubility_unstable(25)
yie = crystal_mass - crystal_mass_final
print(f"Crystal mass = {crystal_mass} kg, yield = {yie} kg, yield% = {yie/crystal_mass*100}%")
```
